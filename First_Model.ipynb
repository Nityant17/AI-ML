```python
import pandas as pd
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error
from sklearn.model_selection import train_test_split
```

```python
# Load training data
train_file_path = '~/Downloads/train.csv'
home_data = pd.read_csv(train_file_path)
```

```python
# Define the Target and Features
y = home_data.SalePrice
features = ['LotArea', 'YearBuilt', '1stFlrSF', '2ndFlrSF', 'FullBath', 'BedroomAbvGr', 'TotRmsAbvGrd','MSSubClass','OverallQual','OverallCond','YearRemodAdd','LowQualFinSF','GrLivArea','HalfBath','KitchenAbvGr','Fireplaces','WoodDeckSF','OpenPorchSF','EnclosedPorch','3SsnPorch','ScreenPorch','PoolArea','MiscVal','MoSold','YrSold']
X = home_data[features]
X.head()
```

```python
# Split into validation and training data
train_X, val_X, train_y, val_y = train_test_split(X, y, random_state=1)
```

```python
# Define a random forest model
rf_model = RandomForestRegressor(random_state=1)
rf_model.fit(train_X, train_y)
rf_val_predictions = rf_model.predict(val_X)
rf_val_mae = mean_absolute_error(rf_val_predictions, val_y)
```

```python
# Find Mean Absolute Error to see accuracy 
print("Validation MAE for Random Forest Model: {:,.0f}".format(rf_val_mae))
```

```python
# Training model on entire training dataset
rf_model_on_full_data = RandomForestRegressor(random_state=1)
rf_model_on_full_data.fit(X,y)
```

```python
# Load testing data
test_data_path = '~/Downloads/test.csv'
test_data = pd.read_csv(test_data_path)
```

```python
# Run the model on testing data
test_X = test_data[features]
test_preds = rf_model_on_full_data.predict(test_X)
print(test_preds)
```

```python
# Loading predictions 
output = pd.DataFrame({'Id': test_data.Id, 'SalePrice': test_preds})
output.head()
# output.to_csv('submission.csv', index=False)  To write the submissions into a csv file
```
